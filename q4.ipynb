```python
import numpy as np
import pandas as pd

# Read the CSV file into a DataFrame
wine = pd.read_csv('WineQT.csv', sep=',')

# Display the first 5 rows of the DataFrame
wine.head()
```

```python
from sklearn.preprocessing import MinMaxScaler

# Step 1: Convert the quality attribute
wine['quality'] = wine['quality'].apply(lambda x: 0 if x <= 6 else 1)

# Step 2: Normalize all other attributes using min-max scaling
scaler = MinMaxScaler()

# Exclude 'Id' and 'quality' from normalization
attributes = wine.columns.difference(['Id', 'quality'])

wine[attributes] = scaler.fit_transform(wine[attributes])

# Display the transformed dataset
wine.head(20)

```
